```python
import pandas as pd
from pandas import Series, DataFrame

import requests
from bs4 import BeautifulSoup
```

```python
def TodayMarket() :

    value = []

    for i in range(1,80) :

        url = 'https://finance.naver.com/sise/sise_market_sum.nhn'  
        site = requests.get(url)
        soup = BeautifulSoup(site.content,"html.parser")

        table = soup.find("tbody")
        trs = table.find_all("tr")[i]  #다음은 2,3,...
        try :
            company_name = trs.find("a",{"class":"tltle"}).text  
        except :
            continue
        today_price = trs.find_all("td",{"class":"number"})[0].text  # 현재가
        related = trs.find_all("td",{"class":"number"})[1].text
        rate = trs.find_all("td",{"class":"number"})[2].text
        total_price = trs.find_all("td",{"class":"number"})[4].text # 시가총액

        a = (company_name)
        b = (today_price)
        c = (related.strip())
        d = (rate.strip())
        e = (total_price)

        url = 'https://search.naver.com/search.naver?sm=top_hty&fbm=0&ie=utf8&query='+a
        site = requests.get(url)
        soup = BeautifulSoup(site.content,"html.parser")

        title = soup.find_all("a",{"class":"_sp_each_title"})[0].text  # 다음은 1,2....
        try :
            source = soup.find_all("span",{"class":"_sp_each_source"})[0].text
        except :
            continue
        f = title
        g = source

        A = [a,b,c,d,e,f,g]
        value.append(A)
        
    RESULT = {}
    for i in range(0,len(value)) :
        co_data = [value[i][0],value[i][1],value[i][2],value[i][3],
                    value[i][4],value[i][5],value[i][6]]
        RESULT[i+1] = co_data
            
    df = pd.DataFrame(RESULT)
    df.index = ['종목명','현재가','전일비','등락률','시가총액','HeadNews','신문사']
    a = df.T
    return a

today = TodayMarket()
```
